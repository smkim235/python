```python
con = sqlite3.connect('example.db')
cur = con.cursor()

query = 'select * from user_score'
rows = cur.execute(query).fetchall()
for row in rows:
    print(row)
    
con.close()
```

```python
con = sqlite3.connect('example.db')
cur = con.cursor()

query = 'select user_score.*, j2.* \
            from user_score \
            inner join user_score j2 on user_score.id = j2.id'
rows = cur.execute(query).fetchall()
for row in rows:
    print(row)
    
con.close()
```

```python
con = sqlite3.connect('example.db')
cur = con.cursor()

query = 'select id, max(score) as m1 \
            from user_score \
            group by id'
rows = cur.execute(query).fetchall()
for row in rows:
    print(row)
    
con.close()
```

```python
con = sqlite3.connect('example.db')
cur = con.cursor()

query = 'select user_score.*, j2.* \
            from user_score \
            inner join \
                (select id, max(score) as m1 from user_score group by id) j2 \
                on user_score.id = j2.id and user_score.score = j2.m1'
rows = cur.execute(query).fetchall()
for row in rows:
    print(row)
    
con.close()
```

```python
con = sqlite3.connect('example.db')
cur = con.cursor()

query = 'select user_score.id, max(user_score.subject), j2.m1 \
            from user_score \
            inner join \
                (select id, max(score) as m1 from user_score group by id) j2 \
                on user_score.id = j2.id and user_score.score = j2.m1\
            group by user_score.id'
rows = cur.execute(query).fetchall()
for row in rows:
    print(row)
    
con.close()
```
