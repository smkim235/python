```python
import numpy as np
```

```python
def argsort(a):
    n = len(a)
    indice = range(0,n)
    
    print 'np:'
    print np.argsort(a)
    
    for i in range(n):
        min_i = i
        for j in range(i+1, n):
            if a[j] < a[min_i]:
                min_i = j
        temp = a[i]
        a[i] = a[min_i]
        a[min_i] = temp
        
        t2 = indice[i]
        indice[i] = indice[min_i]
        indice[min_i] = t2
        
    print 'my arg:'
    print indice
    
    return indice
```

```python
arr = [2,1,3,4]
argsort(arr)

arr = [5,4,3,2,1]
argsort(arr)

arr = [1,3,2,5,4,6]
argsort(arr)

arr = [6,7,1,3,2,4,5]
argsort(arr)
```
