```python
from urlparse import urljoin
page = 'https://www.math.brown.edu'

index = {}
links = []
for a in soup.body.find_all('a'):
    link = a.get('href')
    if not link.startswith('#'):
        link = urljoin(page, link)
        if link.startswith('http'):
            if link not in index:
                links.append(link)
                index[link] = 1


links.sort()
print len(links)
for link in links:
    print link
    
```
