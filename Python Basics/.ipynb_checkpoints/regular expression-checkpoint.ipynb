```python
import re
```

# 1. Basic

```python
# basic api

text = 'abcd12 ABcd34#$'
print re.search('[a-z]+', text).group()
print re.match('[a-z]+', text).group()
print re.findall('[a-z]+', text)
print re.split('[a-z]+', text)
print re.sub('[a-z]+', '_',text)
```

```python
# compile

text = 'abcd12 ABcd34#$'
pattern = '[a-z]+'
r = re.compile(pattern)
r.match(text).group()
```

```python
# group

text = 'abcd12ABcd34#$'
m = re.search('([a-zA-Z]+)[0-9]+([a-zA-Z]+)', text)
print m.group()
print m.group(0)
print m.group(1)
print m.group(2)
print m.group(1,2)
```

# 2. Special character

.

?

*

+


\d
Matches any decimal digit; this is equivalent to the class [0-9].

\D
Matches any non-digit character; this is equivalent to the class [^0-9].

\s
Matches any whitespace character; this is equivalent to the class [ \t\n\r\f\v].

\S
Matches any non-whitespace character; this is equivalent to the class [^ \t\n\r\f\v].

\w
Matches any alphanumeric character; this is equivalent to the class [a-zA-Z0-9_].

\W
Matches any non-alphanumeric character; this is equivalent to the class [^a-zA-Z0-9_].

```python
text = 'abcd#$   12AB cd34#$'
print re.search('\w+', text).group()
print re.sub('[\s]+', '_', text)
print re.search('[^\d]+', text).group()
```

```python
text = 'abbbcd#$'
print re.search('..#', text).group()
print re.search('ab?',text).group()
print re.search('ac?',text).group()
print re.search('ab+',text).group()
print re.search('ab*',text).group()
print re.search('a.+d',text).group()
```

```python
text = 'Titian Fall 2 (2016)'
print re.search('\(\d+\)', text).group()
print re.search('\((\d+)\)', text).group(1)
```

# Applications

## detecting email

```python
text = """Dave dave@google.com
Steve steve@gmail.com
Rob rob@gmail.com
Ryan ryan@yahoo.com
"""
pattern = r'[A-Z0-9._%+-]+@[A-Z0-9.-]+\.[A-Z]{2,4}'
# re.IGNORECASE makes the regex case-insensitive
regex = re.compile(pattern, flags=re.IGNORECASE)

regex.findall(text)
```
