```python
import pandas as pd
```

# 1. Series

```python
arr = pd.Series([1,2,3,4])
print arr
```

```python
arr2 = pd.Series([1,2,3,4], index=['a','b','c','d'], name='Value')
print arr2
```

## operations 

```python
# vector operation
print arr + arr

# concat
arr2 = pd.concat((arr, arr), axis=0)
```

## indexing/subsetting

```python
arr = pd.Series([1,2,3,4], index=['a','b','c','d'])

print arr['b']  # select by index
print arr.loc['b']
print arr.iloc[0]  # select by number index

print arr[['a','d']]  # subset by indices
print arr.iloc[0:3]  # subset by number indices
print arr[[True, False, False, False]]  # subset by boolean
print arr[arr > 3]  # subset by condition
```
