```python
def stream_docs(path):
    with open(path, 'r') as f:
        next(f)	# skip header
        for line in f:
            text, label = line.strip().split(',')
            yield text, label

def get_batch(doc_stream, size):
    docs, y = [], []
    try:
        for _ in range(size):
            text, label = next(doc_stream)
            docs.append(text)
            y.append(label)
    except StopIteration:
        return None, None
    return docs, y
```

```python
doc_stream = stream_docs('./house.csv')
X,y = get_batch(doc_stream, size=10)
```
