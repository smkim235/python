```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

# basic plot

```python
plt.plot([1,2,3,4], [1,2,3,4], 'b+')
plt.show()
```

```python
t = np.arange(0., 5., 0.2)
plt.plot(t, t**3)
plt.show()
```

# multiple plot

```python
t1 = np.arange(0.0, 5.0, 0.2)
plt.plot(t1, t1, 'bx', t1, t1**2, 'r+')
```

# subplots

```python
t1 = np.arange(0.0, 5.0, 0.2)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure(1)
plt.subplot(211)  # grid = 2x1, index = 1
plt.plot(t1, t1, 'bx', t1, t1**2, 'r+')

plt.subplot(212)  # grid = 2x1, index = 2
plt.plot(t2, np.cos(2*np.pi*t2), 'k--')
plt.show()
```

# label setup

```python
plt.plot([1,2,3,4], [1,2,3,4], 'ro')

plt.xlabel('x label')
plt.ylabel('y label')
plt.axis([0, 6, 0, 10])

plt.show()
```

# 3D plot

```python
from mpl_toolkits.mplot3d import axes3d

def randrange(n, vmin, vmax):
    return (vmax - vmin)*np.random.rand(n) + vmin

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
n = 100
for c, m, zl, zh in [('r', 'o', -50, -25), ('b', '^', -30, -5)]:
    xs = randrange(n, 23, 32)
    ys = randrange(n, 0, 100)
    zs = randrange(n, zl, zh)
    ax.scatter(xs, ys, zs, c=c, marker=m)


plt.show()
```

```python
from mpl_toolkits.mplot3d import axes3d

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
X, Y, Z = axes3d.get_test_data(0.05)
ax.plot_wireframe(X, Y, Z, rstride=10, cstride=10)

plt.show()
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
theta = np.linspace(-4 * np.pi, 4 * np.pi, 100)
z = np.linspace(-2, 2, 100)
r = z**2 + 1
x = r * np.sin(theta)
y = r * np.cos(theta)
ax.plot(x, y, z)

plt.show()
```
