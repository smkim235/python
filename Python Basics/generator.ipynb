```python
def get_list():
    ret = []
    for i in range(0,4):
        ret.append(i)
    return ret
```

```python
for x in get_list():
    print(x)
```

```python
def generate_list():
    for i in range(0,4):
        yield i
```

```python
for x in generate_list():
    print(x)
```

```python
gen = generate_list()
print(next(gen))
print(next(gen))
print(next(gen))
print(next(gen))
print(next(gen))
```

```python
gen = generate_list()
next(gen)
next(gen)
for x in gen:
    print(x)
```
