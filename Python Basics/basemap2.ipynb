```python
from mpl_toolkits.basemap import Basemap 
import matplotlib.pyplot as plt
% matplotlib inline
```

# Hatai

```python
lllat=17.25
urlat=20.25
lllon=-75
urlon=-71

m = Basemap(
    projection='stere',
    lon_0=(urlon + lllon) / 2,
    lat_0=(urlat + lllat) / 2,
    llcrnrlat=lllat, 
    urcrnrlat=urlat,
    llcrnrlon=lllon, 
    urcrnrlon=urlon,
    resolution='h')    
m.drawcoastlines()    
m.drawstates()    
m.drawcountries() 
plt.show()
```

# San Francisco

```python
# 37.7749° N, 122.4194° W

pos1 = [-122.55, 37.7]
pos2 = [-122.35, 37.82]
center = [(pos1[0]+pos2[0])/2, (pos1[1]+pos2[1])/2]

m = Basemap(projection='lcc',
            lon_0=center[0],
            lat_0=center[1],
            llcrnrlon=pos1[0],
            llcrnrlat=pos1[1],
            urcrnrlon=pos2[0],
            urcrnrlat=pos2[1],
           resolution='h')

m.drawcoastlines()
m.drawcounties()
plt.show()
```
