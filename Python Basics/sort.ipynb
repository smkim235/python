# 1.implement basic sort

```python
def selection_sort(arr):
    n = len(arr)
    for i in range(0,n-1):
        min_idx = i
        for j in range(i+1,n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        temp = arr[i]
        arr[i] = arr[min_idx]
        arr[min_idx] = temp
    return arr
    
arr2d = [ [3,2,1,4,6,5], [1,3,5,2,4,6], [1,2,3,4,5,6], [6,5,4,3,2,1]]
map(selection_sort, arr2d)
```

```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(0, n-1):
        for j in range(0, n-1-i):
            if arr[j+1] < arr[j]:
                temp = arr[j+1]
                arr[j+1] = arr[j]
                arr[j] = temp
    return arr

arr2d = [ [3,2,1,4,6,5], [1,3,5,2,4,6], [1,2,3,4,5,6], [6,5,4,3,2,1]]
map(bubble_sort, arr2d)
```

```python
def merge_sort(arr):
    n = len(arr)
    if n == 1:
        return arr
        
    half = n/2
    arr1 = merge_sort(arr[0:half])
    arr2 = merge_sort(arr[half:n])
    
    ret = []
    i = 0
    j = 0
    while i < half and j < n-half:
        if arr1[i] < arr2[j]:
            ret.append(arr1[i])
            i += 1
        else :
            ret.append(arr2[j])
            j += 1
                
    if i < half:
        ret = ret + arr1[i:half]
    if j < n-half:
        ret = ret + arr2[j:n-half]
        
    return ret

arr2d = [ [3,2,1,4,6,5], [1,3,5,2,4,6], [1,2,3,4,5,6], [6,5,4,3,2,1]]
map(merge_sort, arr2d)
```

```python
import random

def partition(arr, pivot):
    n = len(arr)
    idx_2nd_par = 0  # index of beginning of the 2nd partition
    
    for i in range(0,n):
        if arr[i] > pivot:
            continue
        else:
            temp = arr[idx_2nd_par]
            arr[idx_2nd_par] = arr[i]
            arr[i] = temp
            idx_2nd_par += 1
    
    return idx_2nd_par

def partition2(arr, pivot):
    n = len(arr)
    i = 0
    j = n-1
    while i <= j:  # '=' is necessary for n=1 case. 
        while i < n and arr[i] < pivot:
            i += 1
        while j > -1 and arr[j] > pivot:
            j -= 1
        
        if i < j:
            temp= arr[i]
            arr[i] = arr[j]
            arr[j] = temp
    return i
    

def quick_sort(arr):
    n = len(arr)
    if n < 2:
        return arr
    
    # random pivot
    pivot_idx = random.randint(0,n-1)
    temp = arr[0]
    arr[0] = arr[pivot_idx]
    arr[pivot_idx] = temp
    
    # we may assume that pivot is at index 0
    pivot = arr[0]
    
    arr2 = arr[1:]
    idx_2nd_par = partition(arr2, pivot)
    return quick_sort(arr2[0:idx_2nd_par]) + [pivot] + quick_sort(arr2[idx_2nd_par:])

arr2d = [ [3,2,1,4,6,5], [1,3,5,2,4,6], [1,2,3,4,5,6], [6,5,4,3,2,1]]
map(quick_sort, arr2d)
```
