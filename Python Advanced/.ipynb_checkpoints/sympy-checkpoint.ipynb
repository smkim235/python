# Higgs

```python
def lie(A,B):
    return A*B - B*A
```

```python
m = Matrix([['p', '0'],['0','1/p']])
phi = Matrix([['0', '0'],['p','0']])
phis = m.inv()*phi.transpose()*m
```

```python
A = Matrix([['a', 'b'],['c','d']])
B = Matrix([['e', 'f'],['g','h']])
```

```python
D1 = Matrix([['a', '0'],['0','b']])
D2 = Matrix([['c', '0'],['0','d']])
```

```python
D1.inv()*phi.transpose()*D1 
```

```python
phis
```

```python
lie(phi, A)
```

```python
lie(phi, lie(phis, A))
```

```python
t = np.linspace(0.001, 0.999, 100)

def myf(t):
    return np.cosh(np.log(t))/np.sinh(np.log(t))/t * t**2

plt.plot(t, myf(t), 'rx')
plt.show()
```
