```python
import numpy as np
import scipy

from scipy import integrate
from scipy.stats import norm
```

```python
# Q: integral of x^2 on [0,1]
```

```python
# Q1. direct math : 1/3
sol1 = 1/3
sol1
```

```python
# 2. numerical integral

integrate.quad(lambda x: x**2, 0, 1)
```

```python
# 3. Monte Carlo integral
# N random samples from [0,1]. 
# vol(Omega) * 1/N * sum f(x_i) -> int(f). 
# error = V/sqrt(N)*sigma_N

N = 10**6 # error=0.00x
points = np.random.random(N)
monte_int = (points**2).sum()/N
print(monte_int)
print(monte_int-sol1)
```

```python
# Q2: X ~ N(0,1). E(|X|)?
```

```python
# 1. direct math : sqrt(2/pi)

sol2 = np.sqrt(2/np.pi)
sol2
```

```python
# 2. numerical integral

num_int = integrate.quad(lambda x: x*norm.pdf(x), 0, 5)
print(2*num_int[0])
print(2*num_int[1])
```

```python
# 3. Monte Carlo

N = 10**6 # error=0.00x
points = np.abs(np.random.randn(N))
monte_sample_mean = points.sum()/N
print(monte_sample_mean)
print(monte_sample_mean-sol2)
```

```python
# 2. Monte Carlo
# Monte Carlo integration for f(x,y) = 1 on 1/4-circle on [0,1)*[0,1)

N = 10**8 # error=0.000x
points_x = np.random.random(N)
points_y = np.random.random(N)
monte_int = (points_x**2 + points_y**2 < 1).sum()/N
sol = monte_int*4
print(sol)
print(sol-np.pi)
```
