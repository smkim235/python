```python
import math

fac = math.factorial
def c(n,r):
    return fac(n) / fac(r) / fac(n-r)
```

```python
c(4,2)
```

```python
c(100,50)
```

```python
#Q3: Out of 2 decks, 7 people receive 5-cards each.  What is the probability that some two people will have 4 aces each.

# sol1 by Dymitro: focus on possible positions of aces:
# c(7,2)*c(5,4)*c(5,4)/c(104,8)


# sol2 by direct count
# index 7 people by 1,2,..7
# all count = c(104,5)*c(99,5)*..
# some 2 people get 4 aces : c(7,2)*c(8,4)*c(96,1)*c(95,1)*..
```

```python
sol1 = c(7,2)*c(5,4)*c(5,4)/c(104,8)
sol1
```

```python
all_count = c(104,5)*c(104-5,5)*c(104-10,5)*c(104-15,5)*c(104-20,5)*c(104-25,5)*c(104-30,5)
target_count = c(7,2)*c(8,4)*c(96,1)*c(95,1)*c(104-10,5)*c(104-15,5)*c(104-20,5)*c(104-25,5)*c(104-30,5)
sol2 = target_count/all_count
sol2
```
